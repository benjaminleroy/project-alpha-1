```python
# Import modules.
from __future__ import (absolute_import,
                        division,
                        print_function,
                        unicode_literals)
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import nibabel as nib
```

```python
# Load BART 4-D data.
path = "/home/oski/s159/project-alpha/ds009/sub013"
img = nib.load(path+"/BOLD/task002_run001/bold.nii.gz")
data = img.get_data()
data.shape
```

```python
# Plot center slice of first volume.
plt.imshow(data[:, :, 16, 0], cmap='gray')
```

```python
# Plot center slice of last volume.
plt.imshow(data[:, :, 16, 183], cmap='gray')
```

```python
# Plot center slice of some volume in the middle.
plt.imshow(data[:, :, 16, 90], cmap='gray')
```

```python
# variance across the final (time) axis
var_vol = np.var(data, axis=3)
plt.imshow(var_vol[:, :, 16], cmap='gray')
```

```python
# Get standard deviation for each volume; then plot the values
# Maybe want to get rid of the first few values.
stds = []
for i in range(data.shape[-1]):
    vol = data[..., i]
    stds.append(np.std(vol))
plt.plot(stds)
```
